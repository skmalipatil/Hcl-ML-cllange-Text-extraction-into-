```python
#defining all the variables
date = ['2018', '2019']
key_words = ['Capital', 'and', 'reserves']
key_words1 = ['Total', 'Shareholder', 'funds']
```

```python
# reading of all lines in the indivisual files 
def tokenizing_all_sentence(file):
    #text_line = []
    with open(file) as f:
        #print(f)
        text_line = []
        for each_line in f:
            text_line.append(each_line)
        
        #file_line_token = []
        line_token = []
        for i in text_line:
            line_token.append(WhitespaceTokenizer().tokenize(i))
        #file_line_token.append(line_token)
        
    return(line_token)
```
